```python
vgg = tf.keras.applications.VGG19(include_top=False, weights='imagenet')

print()
for layer in vgg.layers:
  print(layer.name)
```

```python
content_layers = ['block5_conv2'] 

style_layers = ['block1_conv1',
                'block2_conv1',
                'block3_conv1', 
                'block4_conv1', 
                'block5_conv1']

num_content_layers = len(content_layers)
num_style_layers = len(style_layers)
```

### Create model 

```python
def vgg_layers(vgg,layer_names):
    
    """ Creates a vgg model that returns a list of intermediate output values."""
    # Load our model. Load pretrained VGG, trained on imagenet data
    vgg.trainable = False

    outputs = [vgg.get_layer(name).output for name in layer_names]

    model = tf.keras.Model([vgg.input], outputs)
    return model
```

```python
content_extractor = vgg_layers(vgg,content_layers)
content_outputs = content_extractor(content_image)

#Look at the statistics of each layer's output
for name, output in zip(content_layers, content_outputs):
    print(name)
    print("  shape: ", output.numpy().shape)
    print("  min: ", output.numpy().min())
    print("  max: ", output.numpy().max())
    print("  mean: ", output.numpy().mean())
    print()
```

```python
style_extractor = vgg_layers(vgg,style_layers)
style_outputs = style_extractor(style_image)

#Look at the statistics of each layer's output
for name, output in zip(style_layers, style_outputs):
    print(name)
    print("  shape: ", output.numpy().shape)
    print("  min: ", output.numpy().min())
    print("  max: ", output.numpy().max())
    print("  mean: ", output.numpy().mean())
    print()
```

```python
generated_image = tf.Variable(tf.image.convert_image_dtype(content_image, tf.float32))
noise = tf.random.uniform(tf.shape(generated_image), -0.25, 0.25)
generated_image = tf.add(generated_image, noise)
generated_image = tf.clip_by_value(generated_image, clip_value_min=0.0, clip_value_max=1.0)
generated_image = tf.Variable(generated_image)
#print(generated_image)

print(generated_image.shape)
imshow(generated_image.numpy()[0])
plt.show()
```

```python
def clip_0_1(image):
    return tf.clip_by_value(image, clip_value_min=0.0, clip_value_max=1.0)
```

```python
learning_rate=0.02

style_weight=1e-5
content_weight=1
```

```python
opt = tf.optimizers.Adam(learning_rate=learning_rate, beta_1=0.99, epsilon=1e-1)
```

```python
def get_parameters(A):
    _, n_H, n_W, n_C = A.get_shape().as_list()
    
    return [n_H, n_W, n_C]
```

```python
def gram_matrix(A):
    
    n_H, n_W, n_C = get_parameters(A)
    A = tf.transpose(tf.reshape(A, [-1 , n_C] ))
    gram = tf.matmul(A,tf.transpose(A))
    
    return gram
```

```python
def content_style_loss(generated_image):
    
    generated_content_extractor = vgg_layers(vgg,content_layers)
    generated_content_outputs = generated_content_extractor(generated_image*255)
    #print(generated_content_outputs)
    
    generated_style_extractor = vgg_layers(vgg,style_layers)
    generated_style_outputs = generated_style_extractor(generated_image*255)
    #print(generated_style_outputs)
    
        
        content_loss = tf.add_n([tf.reduce_sum(tf.square(tf.subtract(content_outputs[i],generated_content_outputs[i]))) for i in range(num_content_layers)])
    content_loss *= content_weight / (num_content_layers*2*np.prod(content_outputs[0].get_shape().as_list()))
    
    
    style_loss = tf.add_n([tf.reduce_sum(tf.square(tf.subtract(gram_matrix(style_outputs[i]),gram_matrix(generated_style_outputs[i]))))/ (num_content_layers*2*np.prod(style_outputs[i].get_shape().as_list())) for i in range(num_style_layers)])
    style_loss *= style_weight / (num_style_layers)
    
    loss = style_loss + content_loss 

    return loss
```

```python
loss = content_style_loss(generated_image) 
print(loss)
```

```python
tf.config.experimental_run_functions_eagerly(True)
tf.config.run_functions_eagerly(True)
```

```python
@tf.function()
def train_step(generated_image):
    with tf.GradientTape() as tape:
        #tape.watch(generated_image)
        loss = content_style_loss(generated_image)
        #print(loss)
        #print(generated_image.shape)

    grad = tape.gradient(loss, generated_image)
    opt.apply_gradients([(grad, generated_image)])
    generated_image.assign(clip_0_1(generated_image))
    #print(str(loss) + '\n' )
```

```python
train_step(generated_image)
train_step(generated_image)
train_step(generated_image)
train_step(generated_image)
train_step(generated_image)
train_step(generated_image)
train_step(generated_image)

print(generated_image.shape)
imshow(generated_image[0])
plt.show()
```

```python
epochs = 2501
for i in range(epochs):
    train_step(generated_image)
    if i % 250 == 0:
        print(f"Epoch {i} ")
    if i % 250 == 0:
        imshow(generated_image[0])
        plt.show()
```
